```python
#sort data by age
users_data=(
    users_data
     .sort_values(by="salary",ascending=False)
     .reset_index(drop=True)
)
users_data.index=users_data.index+1
users_data
```

```python
users_data=(
  users_data
    .sort_values(by=["age","salary"],ascending=[True,False])
    .reset_index(drop=True)
    
)
users_data.index=range(1,len(users_data)+1)
#display results
users_data
```
